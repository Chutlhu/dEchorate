## Incemental metrics vs bf

Given the following signal model
$$ \mathbf{c}(f) = [ h_1(f), h_2{f}, \dots, h_I(f) ] s(f)$$
and let assume 1 to be the reference microphones,

We consider the following metris

$$ \mathtt{SNR}_\text{input} = 10 \log_{10} \frac{\mathbb{E}[(h_1 s)^2]}{\sigma^2_{n_1}} $$
$$ \mathtt{SIR}_\text{input} = 10 \log_{10} \frac{\sigma^2_{c_1}}{\sigma^2_{n_1}} $$

Overall performances in terms of:
- Signal to Noise and Revemberant Ratio improvement (SNRR)
    - \[dB\] 
    - (power early speech) / (power diffuse noise + power reverberation)
    - relative metric
    - $\Delta\mathtt{SNRR} = \mathtt{SNRR}_\text{output} - \mathtt{SNRR}_\text{input}$, where
        $$ \mathtt{SNRR}_\text{input} = 10 \log_{10} \left( \frac{\sigma_{h_e \ast s}^2}{\sigma_{h_f \ast s}^2 + \sigma_{n}^2}\right)$$
        where $h_e \ast s$ is the energy of early components of the receiver signal, while $h_f \ast s$ refres to the late reverberation energy.


- speech to reverberation energy modulation ratio  (SRMR)
    - from speechmetrics
    - dimensionless metric
    - absolute metric
    - $\Delta\mathtt{SRMR} = \mathtt{SRMR}_\text{output} - \mathtt{SRMR}_\text{input}$

- Perceptual evaluation of speech quality (PESQ)
    - from speechmetrics
    - dimensionless metric (from 0 to 5)
    - relative metric
    - $\Delta\mathtt{PESQ} = \mathtt{PESQ}_\text{output} - \mathtt{PESQ}_\text{input}$

averaging all the results for:
- all input SNR (0, 10, 20)
- all src position (4 sources)
- all array position (6 positions)
- for \[1,3,5\] dataset (namely RT60) 

```python
df['snr_inc'] = df['snr_out'] - df['snr_in']
df['snrr_inc'] = df['snrr_out'] - df['snrr_in']
df['srmr_inc'] = df['srmr_out'] - df['srmr_in']
df['pesq_inc'] = df['pesq_out'] - df['pesq_in']
```

```python
loc_res = df.loc[(df['dataset'] == 5)]

# gevdlateMVDR' 'rakelateMVDR' 'dpDS' 'gevdMVDR' 'rakeMVDR' 'lateMVDR'
name = {
    'dpDS'     : '$\mathtt{DS}$',
    'MVDR_dp' : '$\mathtt{MVDR}_\mathtt{DP}$',
    'MVDR_rtf' : '$\mathtt{MVDR}_\mathtt{ReTF}$',
    'MVDR_rake'     : '$\mathtt{MVDR}_\mathtt{RAKE}^{*}$',
    'MVDR_dp_late' : '$\mathtt{MVDR}_\mathtt{DP+Late}$',
    'MVDR_rtf_late' : '$\mathtt{MVDR}_\mathtt{ReTF+Late}$',
    'MVDR_rake_late' : '$\mathtt{MVDR}_\mathtt{RAKE+Late}^{*}$',
}
name = {
    'dpDS'     : '$\mathtt{DS}$',
    'MVDR_dp' : '$\mathtt{MVDR-DP}$',
    'MVDR_rtf' : '$\mathtt{MVDR-ReTF}$',
    'MVDR_rake'     : '$\mathtt{MVDR-RAKE*}$',
    'MVDR_dp_late' : '$\mathtt{MVDR-DP-Late}$',
    'MVDR_rtf_late' : '$\mathtt{MVDR-ReTF-Late}$',
    'MVDR_rake_late' : '$\mathtt{MVDR-RAKE-Late*}$',
}

width = 0.7
angle = 15
for key in name.keys():
    loc_res = loc_res.replace(key,    value=name[key])
    print(key)

loc_res = loc_res.replace('real', value='measured')
loc_res = loc_res.replace('synt', value='synthetic')

order_hue = [name[x] for x in name.keys()]
showfliers = False
sns.set(font_scale=1.2)

meanprops={"marker": "+", 
           "markeredgecolor" : "black",
           "markersize": "10"}
# meanprops={}

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))

g1 = sns.boxplot(x='data', y='snrr_inc', hue='bf', data=loc_res, order=['synthetic','measured'],
                 showmeans=True, meanprops=meanprops,
                 hue_order=order_hue, showfliers=showfliers, ax=ax1)
# g1.set(ylabel=r'$\Delta\mathtt{SNRR}$ [dB]', xlabel='')
g1.set(ylabel='', xlabel='')
g1.set_title(r'$\Delta\mathtt{SNRR}$ [dB]')
g1.legend_.remove()

g2 = sns.boxplot(x='data', y='srmr_inc', hue='bf', data=loc_res, order=['synthetic','measured'], 
                 showmeans=True, meanprops=meanprops,
                 hue_order=order_hue, showfliers=showfliers, ax=ax2)
# g2.set(ylabel=r'$\Delta\mathtt{SRMR}$', xlabel='')
g2.set(ylabel='', xlabel='')
g2.set_title(r'$\Delta\mathtt{SRMR}$ [dB]')
g2.legend_.remove()

g3 = sns.boxplot(x='data', y='pesq_inc', hue='bf', data=loc_res, order=['synthetic','measured'],
                 showmeans=True, meanprops=meanprops,
                 hue_order=order_hue, showfliers=showfliers, ax=ax3)
# g3.set(ylabel=r'$\Delta\mathtt{PESQ}$', xlabel='')
g3.set(ylabel='', xlabel='')
g3.set_title(r'$\Delta\mathtt{PESQ}$ [dB]')

plt.legend(bbox_to_anchor=(1.07, .7), loc=2, ncol=1, borderaxespad=0.)
plt.tight_layout()

plt.savefig('../recipes/echo_aware_processing/kowalkzy_results_boxplot.pdf', dpi=300)
plt.savefig('../deliverables/eurasip/figures/kowalkzy_results_boxplot.pdf', dpi=300)
```
